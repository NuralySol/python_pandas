**groupby**
- get groups by column name / category
- get groups by count or by mean

```python
# group students_df by parental edu count, in descending order
# there are six categories, with "some college" having the most occurrences:
parental_edu_grouped_df = students_df.groupby('parental edu').count().sort_values(by='lunch', ascending=False)
```

```python
parental_edu_grouped_df
```

```python
# same as above, but only one of the columns is needed, since values are all the same
parental_edu_grouped_df = students_df.groupby('parental edu').count().sort_values(by='lunch', ascending=False)[['gender']]
parental_edu_grouped_df
```

**Data Visualization with Matplotlib**.

We will delve more deeply into Data Visualization in coming lessons, so consider this a sneak preview
- we will **chart** some of the students data
- **charts** show data on an **x-y coordinate system**
- **bar** chart consists of vertical bars
- **Y-axis** is numeric values
- **X-axis** is categories
- we will make a bar chart from dummy data
- then we will make a bar chart from the students_df **groupby()** data
- **plt.bar()** that takes lists of x-y values as its inputs



```python
# make a bar chart of parental edu counts
# make a list of colors to use for the 6 bars
colors = ["blue", "purple", "orange", "coral", "pink", "darkgreen"]

# make a list of the index names, which are the parental edu categories
edu_categories =  list(parental_edu_grouped_df.index)
print(edu_categories)
# ['some college', "associate's degree", 'high school', 'some high school', "bachelor's degree", "master's degree"]

# the y axis is numeric values, which is for our only number: the count
edu_counts = list(parental_edu_grouped_df['count'])
print(edu_counts) # [226, 222, 196, 179, 118, 59]

# make a bar chart, passing the bar() method the edu_categories and edu_counts:


# show the values on the bars
# fig, ax = plt.subplots()

# ax.bar(edu_counts, edu_counts, 0.75, color = "green")

# # enumerate(list) gives access to index not just item
# for i, item in enumerate(edu_counts):
#     ax.text(item + 3, i + .25, str(edu_counts),
#             color = 'blue', fontweight = 'bold')

plt.bar(edu_categories, edu_counts, color=colors)
plt.xticks(rotation=45)
plt.title('Parental Edu Counts')
plt.ylabel('Number of Students')

plt.show()
```
